```python
def func_to_hsv(image):
    new_image = np.zeros(image.shape)
    for x in range(image.shape[0]):
        for y in range(image.shape[1]):
                r = image[x,y,0]/255
                g = image[x,y,1]/255
                b = image[x,y,2]/255
                maxx = max(r,g,b)
                minn = min(r,g,b)
                delta = maxx - minn

                if delta == 0:
                    h = 0
                elif maxx == r:
                    h = (60 * ((g-b)/delta) + 360) % 360
                elif maxx == g:
                    h = (60 * ((b-r)/delta) + 120) % 360
                elif maxx == b:
                    h = (60 * ((r-g)/delta) + 240) % 360
                v = maxx
                if maxx == 0:
                    s = 0
                else:
                    s = delta/maxx

                new_image[x,y] = [(h/360), (s), (v)]

    return new_image
```

```python
rgb2hsv = func_to_hsv(image)
```

```python
# Task 3. Apply Filters to blur the image and find the one that keeps the most information.
```

```python
import numpy as np
from skimage import io, color, data
import cv2
from skimage.metrics import structural_similarity
```

```python
image_cat = io.imread('cat2.jpg')
```

```python
# Median filter
```

```python
image_grey =cv2.cvtColor(image_cat, cv2.COLOR_BGR2GRAY)
median_image = cv2.medianBlur(image_grey,5)
bilateral_image = cv2.bilateralFilter(image_grey, 2,100,50)
gauss_image = cv2.GaussianBlur(image_grey, (5,5), cv2.BORDER_DEFAULT)
```

```python
ssim_original = structural_similarity(image_grey, image_grey, data_range=image_grey.max() - image_grey.min(), multichannel=True)
ssim_median = structural_similarity(image_grey, median_image, data_range=image_grey.max() - image_grey.min(), multichannel=True)
ssim_bilateral = structural_similarity(image_grey, bilateral_image, data_range=image_grey.max() - image_grey.min(), multichannel=True)
ssim_gauss = structural_similarity(image_grey, gauss_image, data_range=image_grey.max() - image_grey.min(), multichannel=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(12,20))
orig = plt.subplot(1,4,1)
io.imshow(image_grey)
orig.set_title(ssim_original)

median = plt.subplot(1,4,2)
io.imshow(median_image)
median.set_title(ssim_median)

bilateral = plt.subplot(1,4,3)
io.imshow(bilateral_image)
bilateral.set_title(ssim_bilateral)

gauss = plt.subplot(1,4,4)
io.imshow(gauss_image)
gauss.set_title(ssim_gauss)
```

```python
# Task 4. Eye detection application.
```

```python
import cv2

camera = cv2.VideoCapture(0)

eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")

while True:
    ret, frame = camera.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    eyes = eye_cascade.detectMultiScale(gray, 1.2, 5, minSize=(10, 10), flags=cv2.CASCADE_SCALE_IMAGE)

    for (x, y, w, h) in eyes:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 245, 67), 2)

    cv2.imshow("Video Frame", frame)

    if cv2.waitKey(1) == ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```
