```python
# Task 3. Apply Filters to blur the image and find the one that keeps the most information.
```

```python
import numpy as np
from skimage import io, color, data
import cv2
```

```python
image_cat = io.imread('cat2.jpg')
```

```python
median_image = cv2.medianBlur(image_cat,3)
```

```python
gaussian_image = cv2.GaussianBlur(image_cat, (3, 3),  0)
```

```python
bilateral_image = cv2.bilateralFilter(image_cat, 3, 75, 75)
```

```python
from skimage import io
from skimage.metrics import structural_similarity
import cv2
import numpy as np
```

```python
image_cat = io.imread('cat2.jpg')
```

```python
median_image = cv2.resize(median_image, dsize=image_cat.shape[:2])
gaussian_image = cv2.resize(gaussian_image, dsize=image_cat.shape[:2])
bilateral_image = cv2.resize(bilateral_image, dsize=image_cat.shape[:2])
```

```python
ssim_original = structural_similarity(image_cat, image_cat,
                    data_range=image_cat.max() - image_cat.min(), multichannel=True)
```

```python
ssim_median = structural_similarity(image_cat, median_image,
                    data_range=median_image.max() - median_image.min(), multichannel=True)
```

```python
ssim_gaussian = structural_similarity(image_cat, gaussian_image,
                    data_range=gaussian_image.max() - gaussian_image.min(), multichannel=True)
```

```python
ssim_bilateral = structural_similarity(image_cat, bilateral_image,
                    data_range=bilateral_image.max() - bilateral_image.min(), multichannel=True)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,4,1)
ax1.imshow(image_cat)

ax2 = fig.add_subplot(2,4,2)
ax2.imshow(median_image)

ax3 = fig.add_subplot(2,4,3)
ax3.imshow(gaussian_image)

ax4 = fig.add_subplot(2,4,4)
ax4.imshow(bilateral_image)

ax5 = fig.add_subplot(2,4,5)
ax5.imshow(ssim_original)

ax6 = fig.add_subplot(2,4,6)
ax6.imshow(ssim_median)

ax7 = fig.add_subplot(2,4,7)
ax7.imshow(ssim_gaussian)

ax8 = fig.add_subplot(2,4,8)
ax8.imshow(ssim_bilateral)
```

```python
# Task 4. Eye detection application.
```

```python
import cv2

camera = cv2.VideoCapture(0)

eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")

while True:
    ret, frame = camera.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    eyes = eye_cascade.detectMultiScale(gray, 1.2, 5, minSize=(10, 10), flags=cv2.CASCADE_SCALE_IMAGE)

    for (x, y, w, h) in eyes:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 245, 67), 2)

    cv2.imshow("Video Frame", frame)

    if cv2.waitKey(1) == ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```
